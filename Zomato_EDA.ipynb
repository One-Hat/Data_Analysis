```python
df_country= pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
df_country.columns
```

```python
df_final = pd.merge(df,df_country,on = 'Country Code', how='left')
df_final.head(2)
```

```python
df_final.Country.value_counts()
```

```python
country_val=df_final.Country.value_counts().values
```

```python
country_names = df_final.Country.value_counts().index
```

## PIE CHART

```python
plt.pie(country_val[:3],labels=country_names[:3], autopct='%1.2f%%')
```

```python
df_final.columns
```

```python
ratings=df_final.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns = {0:'Rating Count'})
```

```python
ratings
```

## OBServation

```python
## when rating is between 4.5 to 4.9 rating is excellent
## when rating is above 4 and below 4.5 then rating is very god
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (10,6)
sns.barplot(x="Aggregate rating", y = "Rating Count",hue='Rating color',data= ratings)
## also if you want to customise the color of data plots use pallete in the sns , after hue
```

OBs:
1. Not rated count depicted by blue is very high
2. maax number of rating are betwewn 2.5 to 3.4

```python
##Count plot
sns.countplot(x="Rating color",data=ratings)
```
